```python
%load_ext autoreload
%autoreload 2
import test
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(2,1, figsize=(7,6))
for eps in [1,2,3,4]:
    UCNA = test.UCNAprofile(-10,10, epsilon=eps,npoints=10000, zeta=1e12,d=1)
    UCNA.plot_potential(ax=ax[0], derivatives=False, legend=False)
    ax[1].plot(UCNA.x,np.exp(-UCNA.effective_potential(UCNA.x))*UCNA.g, label=str(eps))
    ax[1].plot(UCNA.x, np.ones_like(UCNA.x), '--k')
plt.legend();
# plt.xlim(-2,2)
# plt.ylim(0,10)
```
